### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários
    
    
2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos
    
    
3. Qual o % de funcionários que já fechou algum contrato?<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>

    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 
    
    
4. Calcule o total de contratos que cada área da empresa já fechou


5. Calcule o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

Importação de Módulos e Arquivos

```python
#importando o pandas 
import pandas as pd

#importando os arquivos de Excel e atribuindo-os á uma variável
funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
baseServicoPrestado_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#retirar colunas inúteis para analisar os dados

funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

#exibindo as tabelas
display(clientes_df)
display(funcionarios_df)
display(baseServicoPrestado_df)

```

1 - Folha Salarial 


```python
#Criar uma coluna para obter o salário total somando os respectivas colunas principais: Salario Base, Impostos, Beneficios, VT e VR
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('Total da Folha Salarial Mensal é de R${:,}'.format(sum(funcionarios_df['Salario Total'])))
```

2 - Faturamento da Empresa


```python
#Juntar as 2 colunas principais para fazer o calcúlo de cada dataframe, baseServicoPrestado_df: ID CLiente e Tempo Total de Contrato (Meses)
# e ID Cliente, Valor Contrato Mensal do dataframe Clientes_df
faturamentos_df = baseServicoPrestado_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']])
display(faturamentos_df)
#atribuindo a multiplicação á variável faturamentos_df para obter o resultado, multiplicando Tempo Total de Contrato por Valor Contrato Mensal
faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
print('O faturamento total da empresa foi R${:,}'.format(sum(faturamentos_df['Faturamento Total'])))
```

3- % Funcionários Fecharam Contrato

```python
#Obtendo o tamanho exato da coluna ID Funcionario e tirando as duplicatas da coluna 
qtde_funcionario_fecharamContrato = len(baseServicoPrestado_df['ID Funcionário'].unique())
#obtendo o tamanho/comprimento da coluna ID Funcionarios da dataframe funcionarios_df
qtde_funcionario_total = len(funcionarios_df['ID Funcionário'])
#dividindo os contratos fechados dos funcionários pela quantidade de funcionários
print('Percentual de Funcionários que Fecharam Contratos: {:.2%}'.format(qtde_funcionario_fecharamContrato / qtde_funcionario_total))
```

4 - Qtde Contratos por Área

```python
#Juntando ID Funcionarios da tabela fato BaseServicoPrestato com ID Funcionarios, 'Area' da tabela dimensão Funcionários_df
contratos_area_df = baseServicoPrestado_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')
#abtendo a quantidade de contratos por Area dos Funcionarios da empresa
contratos_area_qtde = contratos_area_df['Area'].value_counts()
print(contratos_area_qtde)
#gerando um gráfico de torta para observar as respectivas areas efetivando os contratos
contratos_area_qtde.plot(kind='pie')
```

5 - Funcionários por Área

```python
#obtendo a contagem de cada area
funcionarios_area = funcionarios_df['Area'].value_counts()
print(funcionarios_area)
#gerando um gráfico de barras demonstrando visualmente o volume de funcionários por área
funcionarios_area.plot(kind='bar')
```

6 - Ticket Médio Mensal

```python
# Obtendo a coluna Valor Contrato Mensal e usando uma função mean() para obter a média do contrato fechado mensalmente
ticket_medio = clientes_df['Valor Contrato Mensal'].mean()
print('O ticket Médio Mensal de contratos mensal é R${:,.2f}'.format(ticket_medio))
```
